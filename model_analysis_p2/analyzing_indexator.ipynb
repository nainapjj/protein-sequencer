```python
%matplotlib inline
import Constants
import analyzePdbVolumes as an
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np

Constants.MEAN_PAIRWISE = 6.73278675
Constants.MIN_PAIRWISE = 8.05876754
Constants.MIN_ACTUAL_VOL = 4
Constants.MAX_ACTUAL_VOL = 20.11906332
```

```python
print "Analyzing 3ZOB"
data_3ZOB = an.volumeErrorsForMotifs("data/3ZOB-one.pdb")
data_3ZOB_hydrophobic = an.volumeErrorsForMotifs("data/3ZOB-one.pdb", filterFunc=an.filterInOnlyAllHydrophobic)

print "Analyzing 1DF4"
data_1DF4 = an.volumeErrorsForMotifs("pdbs/1DF4.pdb")
data_1DF4_hydrophobic = an.volumeErrorsForMotifs("pdbs/1DF4.pdb", filterFunc=an.filterInOnlyAllHydrophobic)

print "Analyzing 1P9I"
data_1P9I = an.volumeErrorsForMotifs("pdbs/1P9I.pdb")
data_1P9I_hydrophobic = an.volumeErrorsForMotifs("pdbs/1P9I.pdb", filterFunc=an.filterInOnlyAllHydrophobic)

print "Analyzing 2RJV"
data_2RJV = an.volumeErrorsForMotifs("pdbs/2RJV.pdb")
data_2RJV_hydrophobic = an.volumeErrorsForMotifs("pdbs/2RJV.pdb", filterFunc=an.filterInOnlyAllHydrophobic)

print "Analyzing 2RJY"
data_2RJY = an.volumeErrorsForMotifs("pdbs/2RJY.pdb")
data_2RJY_hydrophobic = an.volumeErrorsForMotifs("pdbs/2RJY.pdb", filterFunc=an.filterInOnlyAllHydrophobic)
```

```python
print "Looking at 1DF4"

plt.hist(data_1DF4["deltas"], 50)
plt.hist(data_1DF4_hydrophobic["deltas"], 50)
plt.xlabel("Actual Volume - Predicted Volume")
plt.show()

plt.scatter(data_1DF4["aVol"], data_1DF4["eVol"])
plt.xlabel("Actual Volume")
plt.ylabel("Expected Volume")
plt.show()

print "Pearson coeff: %f" % pearsonr(data_1DF4["aVol"], data_1DF4["eVol"])[0]

plt.scatter(data_1DF4_hydrophobic["aVol"], data_1DF4_hydrophobic["eVol"])
plt.xlabel("Actual Volume")
plt.ylabel("Expected Volume")
plt.show()

print "Pearson coeff: %f" % pearsonr(data_1DF4_hydrophobic["aVol"], data_1DF4_hydrophobic["eVol"])[0]
```
